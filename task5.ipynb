```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("sales_data.csv")

# Clean column names
df.columns = df.columns.str.strip()
```

```python

# 1️⃣ Total Revenue by Region
revenue_by_region = df.groupby("Region")["Total Revenue"].sum()
revenue_by_region.plot(kind="bar", title="Total Revenue by Region", color="skyblue")
plt.ylabel("Total Revenue")
plt.xlabel("Region")
plt.show()
```

```python
# 2️⃣ Total Profit by Item Type
profit_by_item = df.groupby("Item Type")["Total Profit"].sum()
profit_by_item.plot(kind="bar", title="Total Profit by Item Type", color="orange")
plt.ylabel("Total Profit")
plt.xlabel("Item Type")
plt.show()
```

```python
# 3️⃣ Units Sold by Country
units_by_country = df.groupby("Country")["Units Sold"].sum()
units_by_country.plot(kind="bar", title="Units Sold by Country", color="green")
plt.ylabel("Units Sold")
plt.xlabel("Country")
plt.show()
```
